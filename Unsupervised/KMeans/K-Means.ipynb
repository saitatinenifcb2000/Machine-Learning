```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("Clustering Shopping_data.csv")
df.shape
```

```python
df.head()
```

```python
data = df.iloc[:, 3:5].values
data[:5]
```

```python
plt.figure(figsize = (10, 7))
plt.scatter(df['Annual Income (k$)'], df["Spending Score (1-100)"])
plt.show()
```

# Elbow-Curve Method/

```python
dis = []
k = range(2, 10)
for i in k:
    model = KMeans(n_clusters = i)
    model.fit(data)
    dis.append(round(model.inertia_, 2)) #inertia_ = WCSS score
    
print("Distortions for all 8 models", dis)
```

```python
plt.figure(figsize = (15, 7))
plt.title("Elbow Curve")
plt.plot(k, dis, 'r*-', markersize = 10)
plt.xlabel("K-Groups")
plt.ylabel("Avg. Distance between datapoints within a Cluster");
```

```python
cluster = KMeans(n_clusters = 5)
model = cluster.fit(data)
clusters = model.labels_
print(len(clusters))
clusters
```

```python
centroids = model.cluster_centers_
centroids
```

```python
plt.scatter(centroids[:, 0], centroids[:, 1], c = "red", s = 250);
```

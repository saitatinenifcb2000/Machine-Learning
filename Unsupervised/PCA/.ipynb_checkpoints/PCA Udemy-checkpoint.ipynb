```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
df=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
stdscaler=StandardScaler()
```

```python
df_std=pd.DataFrame(stdscaler.fit_transform(df),columns=df.columns)
```

```python
df_std
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=10)
```

```python
model=pca.fit(df_std)
```

```python
model
```

```python
print('Explained variance for each PCA',model.explained_variance_ratio_)
```

```python
print('Variance explained % is',sum(model.explained_variance_ratio_)*100)
```

```python
df_std.shape
```

```python
x_pca=pca.transform(df_std)
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'])
plt.xlabel('First PCA')
plt.ylabel('Second PCA')
plt.show()
```

```python
pca.components_
```

```python
df_comp=pd.DataFrame(pca.components_,columns=cancer['feature_names'])
```

```python
df_comp
```

```python
sns.heatmap(df_comp)
```

```python
def display_correlation(df) :
    # Compute the correlation matrix
    corr = df.corr()

    # Generate a mask for the upper triangle
    mask = np.zeros_like(corr, dtype=np.bool)
    mask[np.triu_indices_from(mask)] = True

    # Set up the matplotlib figure
    f, ax = plt.subplots(figsize=(11, 9))

    # Generate a custom diverging colormap
    cmap = sns.diverging_palette(220, 10, as_cmap=True)

    # Draw the heatmap with the mask and correct aspect ratio
    g6 = sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.6, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
    g6.set_title("Correlation among columns")
    fig = g6.get_figure()
    fig.savefig("Correlation.png")
    
display_correlation(result)
```

```python
## Removing missing values
# function for displaying variable details - categorical
def view_categorical_columns(col, missing, result):
    print ('column : %s\n'%(col))
    print (result[col].head(), '\n')
    print ('unique values :' %(result[col].unique()))
    print (result[col].value_counts(), '\n')
    print ('total : %d\n' %(len(result[col])))
    print ('missing : %d\n' %(result[col].loc[result[col] == missing].count()))
    print ('na or null : %d\n' %(result[col].isnull().sum()))
    print ('-------------------\n')
```

```python
# function for displaying variable details - numerical
def view_numeric_columns(col, missing, result):
    print ('column : %s\n'%(col))
    print (result[col].head(), '\n')
    print ('total : %d\n' %(len(result[col])) )
    print ('missing : %d\n' %(result[col].loc[result[col] == missing].count()) )
    print ('na or null : %d\n' %(result[col].isnull().sum()))
    print ('-------------------\n')
```

```python
def  view_missing_columns(result, isRemoved):
    ## Categorical

    # InsuredGender ==> 30 na ==>
    view_categorical_columns('InsuredGender', 'nan', result)
    # TypeOfCollission ==> 5162 missing ==>
    view_categorical_columns('TypeOfCollission', '?', result)
    # Witnesses ==> 46 ==>
    view_categorical_columns('Witnesses', 'MISSINGVALUE', result)
    if not isRemoved :
        # PropertyDamage ==> 10459 missing ==> 
        view_categorical_columns('PropertyDamage', '?', result)
        # PoliceReport ==> 9805 missing ==>
        view_categorical_columns('PoliceReport', '?', result)
        # Country
        view_categorical_columns('Country', 'MISSINGVALUE', result)
        # VehicleMake
        view_categorical_columns('VehicleMake', 'MISSINGVALUE', result)

    print('-------------------\n')

    ## Numerical

    # PolicyAnnualPremium ==> 141 missing ==>
    view_numeric_columns('PolicyAnnualPremium', -1, result)
    # IncidentTime ==> 31 missing ==>
    view_numeric_columns('IncidentTime', -5, result)
    # AmountOfTotalClaim ==> 50 missing ==>
    view_numeric_columns('AmountOfTotalClaim', 'MISSEDDATA', result)
    
view_missing_columns(result, False)
```

```python
def preprocess_further(result2, isTrain):
    
    # Splitting zipcodes
    result2['Insured_Region'] = result2.InsuredZipCode//10000
    result2['Insured_SubRegion'] = result2.InsuredZipCode//100000
    result2 = result2.drop(['InsuredZipCode'], axis = 1)
    
    # Removing unique identifier columns
    result2 = result2[result2.columns.difference(['CustomerID', 'VehicleID', 'InsurancePolicyNumber'])]
    
    # Converting response variable to 1 and 0 format
    if isTrain :
        result2['ReportedFraud'] = (result2['ReportedFraud'] =='Y').astype(int)
    
    # Removing location temporarily
    result2 = result2.drop(['IncidentAddress'], axis = 1)
    
    # PropertyDamage - high NA values
    #result2 = result2.drop(['PropertyDamage'], axis = 1)
    result2.loc[result2['PropertyDamage'].isnull(), 'PropertyDamage'] = 'Unknown'
    
    # PoliceReport - high NA values
    #result2 = result2.drop(['PoliceReport'], axis = 1)
    result2.loc[result2['PoliceReport'].isnull(), 'PoliceReport'] = 'Unknown'
    
    # Country - 99% of value is India
    result2 = result2.drop(['Country'], axis = 1)
    
    # AmountOfVehicleDamage - High Correlation
    #result2 = result2.drop(['AmountOfVehicleDamage'], axis = 1)
    
    # Working with date variables and creating new variables out of them.
    date_cols = ['DateOfIncident', 'DateOfPolicyCoverage']

    result2['DateOfIncident'] = pd.to_datetime(result2['DateOfIncident'])
    result2['DateOfPolicyCoverage'] = pd.to_datetime(result2['DateOfPolicyCoverage'])

    result2['MonthOfIncident'] = result2['DateOfIncident'].dt.month

    result2['YearOfIncident'] = result2['DateOfIncident'].dt.year
    result2['YearOfPolicyCoverage'] = result2['DateOfPolicyCoverage'].dt.year
    
    # DateOfIncident - DateOfPolicyCoverage
    result2['PolicyCoverage_Days'] = result2['DateOfIncident'] - result2['DateOfPolicyCoverage']
    result2.PolicyCoverage_Days = result2.PolicyCoverage_Days.dt.days
    
    # Dropping date columns
    result2 = result2.drop(['DateOfIncident', 'DateOfPolicyCoverage'], axis = 1)
    
    # YearOfIncident - YearOfPolicyCoverage
    result2['PolicyCoverage_Years'] = result2['YearOfIncident'] - result2['YearOfPolicyCoverage']
    result2.PolicyCoverage_Years = result2.PolicyCoverage_Years

    # VehicleYOM - YearOfIncident
    result2['VehicleYOM'] = pd.to_datetime(result2['VehicleYOM']).dt.year
    #result2['VehicleYOM'] = pd.to_datetime(result2['VehicleYOM'])
    result2['Depreciation_Years'] = result2['YearOfIncident'] - result2['VehicleYOM']
    result2 = result2.drop(['YearOfIncident'], axis = 1)
    
    # Drop VehicleMake
    result2 = result2.drop(['VehicleMake'], axis = 1)
     
    # Drop VehicleModel
    result2 = result2.drop(['VehicleModel'], axis = 1)
    
    # Expanding Policy_CombinedSingleLimit
    result2[['Liability_PerInjured','Liability_TotalInjured']] = result2['Policy_CombinedSingleLimit'].str.split('/',expand=True)
    result2[['Liability_PerInjured','Liability_TotalInjured']] = result2[['Liability_PerInjured','Liability_TotalInjured']].apply(pd.to_numeric, errors='coerce', axis=1)

    result2['Liability_MaximumCover'] = (result2['Liability_TotalInjured'] / result2['Liability_PerInjured']).round()
    result2.Liability_MaximumCover.value_counts()
    
    # Liability_MaximumCover - BodilyInjuries
    result2['Liability_PossibleCover'] = result2['Liability_MaximumCover'] - result2['BodilyInjuries']
    
    result2 = result2.drop(['Policy_CombinedSingleLimit'], axis = 1)
    
    # Transforming claim amounts to ratios
    result2[['AmountOfTotalClaim']] = result2[['AmountOfTotalClaim']].apply(pd.to_numeric, errors='coerce', axis=1)
    result2['RatioPropertyClaim'] = result2['AmountOfPropertyClaim'] / result2['AmountOfTotalClaim']
    result2['RatioInjuryClaim'] = result2['AmountOfInjuryClaim'] / result2['AmountOfTotalClaim']
    result2['RatioVehicleDamage'] = result2['AmountOfVehicleDamage'] / result2['AmountOfTotalClaim']
    
    # (CaptialGains + CapitalLoss) / 1000 => scaling to 1000s
    result2['Finance_Balance'] = (result2['CapitalGains'] + result2['CapitalLoss']) #/1000
    result2 = result2.drop(['CapitalGains', 'CapitalLoss'], axis = 1)
    
    # Scaling UmbrellaLimit, AmountOfTotalClaim (all amounts of claim), Finance_Balance to 1000s
    result2['UmbrellaLimit'] = result2['UmbrellaLimit'] #/ 1000
    result2['AmountOfTotalClaim'] = result2['AmountOfTotalClaim'] #/ 1000
    result2['Finance_Balance'] = result2['Finance_Balance'] #/ 1000
    
    result2['AmountOfPropertyClaim'] = result2['AmountOfPropertyClaim'] #/ 1000
    result2['AmountOfInjuryClaim'] = result2['AmountOfInjuryClaim'] #/ 1000
    result2['AmountOfVehicleDamage'] = result2['AmountOfVehicleDamage'] #/ 1000
    
    print (result2.dtypes)
    return result2
    
result2 = preprocess_further(result, True)
    
#result2.IncidentTime = result2.IncidentTime.fillna(-5).astype(int)
#fill_missed_values('IncidentTime', -5, result2)
#view_categorical_columns('IncidentTime', -1, result2)
```

```python
def to_save(col_2, one, zero):
    col_1 = outcome_test['CustomerID']
    to_save_result = pd.concat([col_1,col_2], axis=1)
    to_save_result['ReportedFraud'] = to_save_result['predict']
    to_save_result = to_save_result.drop(['predict'], axis = 1)
    to_save_result.loc[to_save_result['ReportedFraud'] == one, 'ReportedFraud'] = 'Y'
    to_save_result.loc[to_save_result['ReportedFraud'] == zero, 'ReportedFraud'] = 'N'

    return to_save_result
```

```python
def plot_perf(grid, test):
    print('best')
    for index, model in enumerate(grid.models):
        print ('model',index)
        print ('id',model.model_id)
        perf_grid = model.model_performance(test)
        print ('gini:',perf_grid.gini())
        print ('auc:',perf_grid.auc())
        print ('--------\n\n')
```

```python
# Log transformation
def log_transform(df, col):
    if( min(df[col])  < 0):
        transformed = np.log1p(df[[col]] - min(df[col]))
    else:
        transformed = np.log1p(df[[col]])
    return transformed


# Log transformation
def sqrt_transform(df, col):
    if( min(df[col])  < 0):
        transformed = np.sqrt(df[[col]] - min(df[col]) + 1)
    else:
        transformed = np.sqrt(df[[col]])
    return transformed

# Log transformation
def n_pow_transform(df, col, n):
    if( min(df[col])  < 0):
        transformed = (np.log(df[[col]] - min(df[col]) + 1)).pow(n)
    else:
        transformed = (np.log(df[[col]])).pow(n) #.pow(1./n)
    return transformed
```

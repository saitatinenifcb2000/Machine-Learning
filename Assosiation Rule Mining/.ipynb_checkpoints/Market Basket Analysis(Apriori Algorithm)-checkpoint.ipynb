```python
retail_data=pd.read_excel('Online Retail.xlsx')
```

```python
retail_data.head()
```

```python
retail_data.shape
```

```python
retail_data['Description']=retail_data['Description'].str.strip()
```

```python
#remove spaces from beginning and end
```

```python
retail_data.isnull().sum()
```

```python
retail_data['InvoiceNo'].value_counts()
```

```python
retail_data.dropna(axis=0,subset=['InvoiceNo'],inplace=True)
```

```python
retail_data['InvoiceNo']=retail_data['InvoiceNo'].astype('str')
```

```python
#converting Invoice number into String type.
```

```python
retail_data=retail_data[~retail_data['InvoiceNo'].str.contains('C')]
```

```python
#Removing the Invoice No with C.
```

```python
retail_data['InvoiceNo'].unique()
```

```python
retail_data.shape
```

```python
retail_data['Country'].value_counts()
```

```python
retail_data.head()
```

```python
#Separating tractations for United Kingdom only
```

```python
mybasket=retail_data[retail_data['Country']== ('United Kingdom')].groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo')
```

```python
mybasket.head()
```

```python
def my_encode_units(x):
    if x<=0:
        return 0
    if x>=1:
        return 1
```

```python
mybasket_sets=mybasket.applymap(my_encode_units)
```

```python
mybasket_sets.head()
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
myfrequent_itemsets=apriori(mybasket_sets,min_support=0.075,use_colnames=True)
```

```python
myfrequent_itemsets
```

```python
my_rules=association_rules(myfrequent_itemsets,metric='lift',min_threshold=1)
```

```python
my_rules.shape
```

```python
my_rules.sort_values(by='lift',ascending=False)
```

```python
my_rules.sort_values(by='lift',ascending=False)[:5]
```

```python
#Making Recommendations
```

```python
my_rules[(my_rules['lift']>=13)]
```

```python
my_rules[(my_rules['lift']>=13) & (my_rules['confidence']>=0.8)]
```
